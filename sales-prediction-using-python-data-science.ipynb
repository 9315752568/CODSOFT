# USING LINEAR REGRESSION 
****

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(dataset[['TV']],dataset[['Sales']],test_size=0.1,random_state=20)
print(X_train)
```

```python
print(X_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
res=model.predict(X_test)
res
```

```python
plt.plot(res)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
plt.scatter(X_test,y_test)
plt.plot(X_test,7.01122865+0.05565261* X_test,'r')
plt.show()
```
